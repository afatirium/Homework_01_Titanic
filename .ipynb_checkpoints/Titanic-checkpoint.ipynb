```python
# Perform chi-squared test for each categorical variable
print("Chi-squared test results for categorical variables:")
for feature in categorical_features:
    contingency_table = pd.crosstab(X_train[feature], y_train)
    chi2, p_val, _, _ = chi2_contingency(contingency_table)
    print(f"Chi-squared test for {feature}: p-value = {p_val:.5f}")
```

The chi-squared test results for the 'Sex' and 'Pclass' variables indicate highly significant associations with the target variable 'Survived'. Let's interpret these results:

1. **Chi-squared test for 'Sex':**
   - **p-value = 0.00000**: The extremely low p-value (very close to zero) suggests a highly significant association between the 'Sex' of a passenger and their likelihood of survival.
   - In other words, the gender ('Sex') of a passenger is strongly related to whether they survived or not on the Titanic. Historical data supports this, showing a higher survival rate among females compared to males.

2. **Chi-squared test for 'Pclass':**
   - **p-value = 0.00000**: Again, the extremely low p-value indicates a highly significant association between the passenger class ('Pclass') and the likelihood of survival.
   - This implies that the class a passenger was in ('Pclass') significantly influenced their chances of survival. For example, passengers in higher classes (lower 'Pclass' numbers) had a better chance of survival compared to those in lower classes.

## 4.1. Modelling


```python
# Define features (X) and target (y) for training data
X_train = df_dataset1.drop('Survived', axis=1)
y_train = df_dataset1['Survived']

# Define features (X) and target (y) for test data
X_test = df_test1.drop('Survived', axis=1)
y_test = df_test1['Survived']
```

```python
X_train.head()
```

```python

# Create and train the logistic regression model with L2 regularization turned off
log_reg = LogisticRegression(penalty='none', random_state=123)
log_reg.fit(X_train, y_train)

# Predict probabilities for the test set
y_prob = log_reg.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

# Compute PR curve
precision, recall, _ = precision_recall_curve(y_test, y_prob)
average_precision = average_precision_score(y_test, y_prob)

# Plot ROC curve
plt.figure(figsize=(10, 5))
ax1 = plt.subplot(1, 2, 1)
plot_roc_curve(log_reg, X_test, y_test, ax=ax1)
ax1.set_title('ROC Curve (AUC = {:.2f})'.format(roc_auc))

# Plot PR curve
ax2 = plt.subplot(1, 2, 2)
plot_precision_recall_curve(log_reg, X_test, y_test, ax=ax2)
ax2.set_title('Precision-Recall Curve (Avg Precision = {:.2f})'.format(average_precision))

plt.tight_layout()
plt.show()
```

# Task 5:

Use cross-validation of your choice (if your sample is unbalanced, use cross-validation with stratified, e.g. https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedKFold.html).

```python

# Define features (X) and target (y)
X = df_dataset1.drop('Survived', axis=1)
y = df_dataset1['Survived']

# Initialize StratifiedKFold with 5 folds
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

# Perform cross-validation and obtain predicted probabilities
y_prob_cv = cross_val_predict(log_reg, X, y, cv=stratified_kfold, method='predict_proba')[:, 1]

# Compute AUC and average precision
auc_cv = roc_auc_score(y, y_prob_cv)
avg_precision_cv = average_precision_score(y, y_prob_cv)

print("Cross-validated AUC: {:.2f}".format(auc_cv))
print("Cross-validated Average Precision: {:.2f}".format(avg_precision_cv))
```

The cross-validated AUC (Area Under the Curve) of 0.85 and cross-validated Average Precision of 0.81 indicate that the logistic regression model is performing well in distinguishing between the positive and negative classes in your dataset. These metrics suggest a strong predictive capability of the model.

- AUC of 0.85 indicates a good ability to discriminate between positive and negative samples. The higher the AUC, the better the model is at distinguishing between survival and non-survival cases.

- Average Precision of 0.81 reflects the trade-off between precision and recall, which is important, especially in imbalanced datasets. A higher average precision indicates that the model is able to identify a substantial number of actual positive cases.

Overall, these metrics suggest a promising performance of the logistic regression model on the Titanic dataset. # Task: 6

Check whether regularization will be helpful in your problem - use three models of regularized logistic regression (ridge, lasso and elastic net). You can use the LogisticRegression object for this (https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression) - note that you have three hyperparameters at your disposal: a) penalty - Specify the norm of the penalty; b) C - Inverse of regularization strength; c) l1_ratio - this remark only makes sense if you are using elastic net. Try playing with these hyperparameters.

```python

# Regularization strengths to try (you can adjust these)
Cs = [0.1, 1, 10]

# Perform logistic regression with different penalties
penalties = ['l2', 'l1', 'elasticnet']
for penalty in penalties:
    if penalty == 'elasticnet':
        # For Elastic Net, we'll manually vary l1_ratio
        l1_ratios = [0.1, 0.5, 0.9]
        for l1_ratio in l1_ratios:
            model = LogisticRegression(penalty=penalty, C=1.0 / Cs[0], solver='saga', l1_ratio=l1_ratio, random_state=123)
            model.fit(X, y)
            print(f"Regularization: {penalty}, l1_ratio: {l1_ratio}")
            print("Train AUC:", roc_auc_score(y, model.predict_proba(X)[:, 1]))
            print("Train Average Precision:", average_precision_score(y, model.predict_proba(X)[:, 1]))
            print("---")
    else:
        model = LogisticRegressionCV(penalty=penalty, Cs=Cs, cv=5, solver='saga', random_state=123)
        model.fit(X, y)
        print(f"Regularization: {penalty}")
        print("Best C:", 1.0 / model.C_[0])  # Inverse of the best C
        print("Train AUC:", roc_auc_score(y, model.predict_proba(X)[:, 1]))
        print("Train Average Precision:", average_precision_score(y, model.predict_proba(X)[:, 1]))
        print("---")
```


1. **L2 Regularization (Ridge):** L2 regularization, also known as Ridge regularization, is a method that adds "squared magnitude" of coefficient weights to the cost function. A lower value of C (0.1 in this case) indicates stronger regularization. The AUC (0.812) and Average Precision (0.709) indicate that the model performs reasonably well in distinguishing between survivors and non-survivors.

2. **L1 Regularization (Lasso):** L1 regularization, or Lasso regularization, adds the "absolute magnitude" of coefficient weights to the cost function. The selected best C (0.1) suggests stronger regularization. The AUC (0.809) and Average Precision (0.706) also indicate good predictive performance of the model.

3. **Elastic Net (l1_ratio: 0.1, 0.5, 0.9):** Elastic Net combines L1 and L2 regularizations. It's observed that varying the l1_ratio doesn't significantly affect the model's performance in this case, suggesting that a mix of L1 and L2 regularization (Elastic Net) doesn't provide substantial improvements over Ridge and Lasso regularization for this dataset.

Use your chosen hyperparameter tuning method for regularized logistic regression. Checkon the out-of-sample test set whether your results have improved.

```python
# Best C values from the previous analysis
best_C_ridge = 0.1
best_C_lasso = 0.1

# Train models with the best C values
ridge_model = LogisticRegression(penalty='l2', C=best_C_ridge, solver='saga', random_state=123)
lasso_model = LogisticRegression(penalty='l1', C=best_C_lasso, solver='saga', random_state=123)

# Define features and target for the test set
X_test = df_test1.drop('Survived', axis=1)
y_test = df_test1['Survived']

# Train the models on the training data
ridge_model.fit(X, y)
lasso_model.fit(X, y)

# Evaluate on the test set
test_auc_ridge = roc_auc_score(y_test, ridge_model.predict_proba(X_test)[:, 1])
test_avg_precision_ridge = average_precision_score(y_test, ridge_model.predict_proba(X_test)[:, 1])

test_auc_lasso = roc_auc_score(y_test, lasso_model.predict_proba(X_test)[:, 1])
test_avg_precision_lasso = average_precision_score(y_test, lasso_model.predict_proba(X_test)[:, 1])

print("L2 Regularization (Ridge) - Test AUC:", test_auc_ridge)
print("L2 Regularization (Ridge) - Test Average Precision:", test_avg_precision_ridge)

print("L1 Regularization (Lasso) - Test AUC:", test_auc_lasso)
print("L1 Regularization (Lasso) - Test Average Precision:", test_avg_precision_lasso)
```

Both models (Ridge and Lasso) perform relatively well on the unseen test data. The AUC and Average Precision values are reasonable, indicating that the models are able to generalize to new, unseen examples.

Comparing the two regularization techniques, the Ridge model slightly outperforms the Lasso model in terms of both AUC and Average Precision on the test set. However, both models perform quite comparably.

```python
# Best C value for L2 (Ridge) regularization from the previous analysis
best_C_ridge = 0.1

# Train the Ridge model using the entire dataset
final_model = LogisticRegression(penalty='l2', C=best_C_ridge, solver='saga', random_state=123)
final_model.fit(X, y)

# Save the final model for future use
import joblib
joblib.dump(final_model, 'final_logistic_regression_model.pkl')
```

```python
# Load the final model
final_model = joblib.load('final_logistic_regression_model.pkl')

# Predict on the test data
y_pred = final_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Test Accuracy:", accuracy)
print("Test Precision:", precision)
print("Test Recall:", recall)
print("Test F1 Score:", f1)
```
